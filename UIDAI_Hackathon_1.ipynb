```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate total enrollments for each district
df1['total_enrollments'] = df1['age_0_5'] + df1['age_5_17'] + df1['age_18_greater']

# Group by district and sum total enrollments, then get the top 10
top_10_districts = df1.groupby('district')['total_enrollments'].sum().nlargest(10).reset_index()

# Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x='total_enrollments', y='district', data=top_10_districts, palette='viridis', hue='district', legend=False)
plt.title('Top 10 Districts by Total Enrollments')
plt.xlabel('Total Enrollments')
plt.ylabel('District')
plt.tight_layout()
plt.show()
```

```python
df1['percentage_0_5'] = (df1['age_0_5'] / df1['total_enrollments']) * 100
df1['percentage_5_17'] = (df1['age_5_17'] / df1['total_enrollments']) * 100
df1['percentage_18_greater'] = (df1['age_18_greater'] / df1['total_enrollments']) * 100

print(df1[['age_0_5', 'age_5_17', 'age_18_greater', 'total_enrollments', 'percentage_0_5', 'percentage_5_17', 'percentage_18_greater']].head())
```

```python
df_time_series = df1.groupby('date')[['percentage_0_5', 'percentage_5_17', 'percentage_18_greater']].mean().reset_index()

print(df_time_series.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Melt the DataFrame to long format for Seaborn lineplot
df_melted = df_time_series.melt(id_vars=['date'],
                                value_vars=['percentage_0_5', 'percentage_5_17', 'percentage_18_greater'],
                                var_name='age_group',
                                value_name='percentage')

# Map age group column names for better labels in the plot
df_melted['age_group'] = df_melted['age_group'].map({
    'percentage_0_5': 'Age 0-5',
    'percentage_5_17': 'Age 5-17',
    'percentage_18_greater': 'Age 18+'
})

# Create the line chart
plt.figure(figsize=(15, 8))
sns.lineplot(x='date', y='percentage', hue='age_group', data=df_melted, marker='o', linewidth=2.5)

plt.title('Percentage Contribution of Age Groups Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Percentage Contribution', fontsize=12)
plt.legend(title='Age Group', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
